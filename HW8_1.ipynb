```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 讀入大專校院圖書館統計

```python
df=pd.read_csv("http://stats.moe.gov.tw/files/detail/107/107_library2.csv")
```

```python
df.head()
```

# 算出館藏資源

```python
df["館藏資源"] = df.loc[:,"中文圖書收藏冊數(冊)-總類":"視聽資料(件)"].sum(axis=1)
```

```python
df.head()
```

# 以館藏資源為主、全年圖書借閱人次為輔找出前五高的學校

```python
df.sort_values(by=["館藏資源","全年圖書(含視聽資料及其他館藏)借閱人次(人次)"], ascending=False).head()
```

# 以全年圖書借閱人次為主、館藏資源為輔找出前五高的學校

```python
df.sort_values(by=["全年圖書(含視聽資料及其他館藏)借閱人次(人次)","館藏資源"], ascending=False).head()
```

# 找出"書本"和"期刊"數位化最高的學校
## 首先用sum算出紙本"書"和"期刊"的數量

```python
df["書(紙本)資源"] = df.loc[:,"中文圖書收藏冊數(冊)-總類":"外文圖書收藏冊數(冊)"].sum(axis=1)
```

```python
df.head()
```

```python
df["期刊(紙本)資源"] = df.loc[:, ["期刊合訂本(未以圖書編目)(冊)", "報紙(種)","中.日文期刊(種)","西文期刊(種)"]].sum(axis=1)
```

```python
df.head()
```

# 因為列的種類太多，把細分的書本跟期刊種類類別刪掉

```python
df.drop(labels=["中文圖書收藏冊數(冊)-總類","中文圖書收藏冊數(冊)-哲學類","中文圖書收藏冊數(冊)-宗教類","中文圖書收藏冊數(冊)-科學類","中文圖書收藏冊數(冊)-應用科學類","中文圖書收藏冊數(冊)-社會科學類","中文圖書收藏冊數(冊)-史地類","中文圖書收藏冊數(冊)-語言文學類","中文圖書收藏冊數(冊)-藝術類","外文圖書收藏冊數(冊)","期刊合訂本(未以圖書編目)(冊)","報紙(種)","西文期刊(種)","中.日文期刊(種)"],axis=1,inplace = True)
```

```python
df.head()
```

# 算出"期刊數位化程度"和"書本數位化程度"

```python
df['期刊數位化程度'] = df.apply(lambda x: x['電子期刊(種)'] / x['期刊(紙本)資源'], axis=1)
```

```python
df.head()
```

```python
df['書本數位化程度'] = df.apply(lambda x: x['電子書(種)'] / x['書(紙本)資源'], axis=1)
```

```python
df.head()
```

# 依據"書本數位化程度"與"期刊數位化程度"排出前五名的學校

```python
df.sort_values(by="書本數位化程度", ascending=False).head()
```

```python
df.sort_values(by="期刊數位化程度", ascending=False).head()
```

```python
df.sort_values(by=["書本數位化程度","期刊數位化程度"], ascending=False).head()
```
