```python
""
with open("/Users/Cyh/Desktop/1021.txt", "r", encoding="utf-8") as f: 
    text=f.read().replace("\n", "").replace ("\3000", "")
wordlist= jieba.cut(text, cut_all = False)
wl=" ".join(wordlist)
wc = WordCloud(font_path="/Users/Cyh/Desktop/msjh_0.ttf", background_color="white", max_words = 200)
wc.generate(wl)
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(10,6), dpi = 100)
plt.show()
```

```python
""
with open("/Users/Cyh/Desktop/1021.txt", "r", encoding="utf-8") as f: 
    text=f.read().replace("\n", "").replace ("\3000", "")
wordlist= jieba.cut(text, cut_all = False)
words = pseg.cut(text)
wa=[]
for w in words:
    if w.flag in ["n","v"]: 
        wa.append(w.word)
for i in wa:
    wl = wl +" " + i
maskf = np.array(Image.open("/Users/Cyh/Desktop/.jpg"))
wc = WordCloud(font_path="/Users/Cyh/Desktop/TaipeiSansTCBeta-Bold.ttf", background_color="white", max_words = 200,mask=maskf , contour_width=1, contour_color="white")
wc.generate(wl)
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(10,6), dpi = 100)
plt.show()
```

```python
""
jieba.analyse.TFIDF(idf_path=None)
with open("/Users/Cyh/Desktop/1021.txt", "r", encoding="utf-8") as f: 
    text=f.read().replace("\n", "").replace ("\3000", "")
seglist = jieba.cut(text, cut_all=False)
a=jieba.analyse.extract_tags(text, topK=20, withWeight=False, allowPOS=())
print(a)
dic={}
for i in seglist:
    if i in dic and i in a:
        dic[i]+=1
    elif i in a and i not in dic:
        dic[i]=1
print(dic)

with open("q1021-3", "a", encoding ="cp950") as f: 
    f.write("word,count\n")
```
